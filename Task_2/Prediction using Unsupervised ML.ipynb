### Step:3 Finding optimum number of clustering using k-means clustering algorithm

```python
x = data.iloc[:,[0,1,2,3]].values

from sklearn.cluster import KMeans
wcss = []
for i in range(1,6):
    kmeans = KMeans(init="k-means++", n_clusters=i, n_init=10, max_iter=300, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    print('k:', i, "wcss:",  kmeans.inertia_)
```

```python
# Plotting the results into a line graph,
# allowing us to observe 'The elbow'

plt.figure(figsize=(10,5))

plt.plot(3,78.94084142614601,color='black',marker='o',markersize=10)

plt.plot(range(1,6),wcss,marker='o')

plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Within cluster sum of squares')
plt.show()
```

<b>Elbow method:</b><br>In cluster analysis, the elbow method is a heuristic used in determining the number of clusters in a data set. The method consists of plotting the explained variation as a function of the number of clusters, and picking the elbow of the curve as the number of clusters to use.

We can observe that the “elbow” is the number 3 which is optimal for this case. Now we can run a K-Means using as n_clusters the number 3.

```python
wcss
```

### Step:4 Initializing K-means with the optimum number of cluster & Predicting the values

```python
#fitting the Kmean in the data
kmeans = KMeans(init = 'k-means++',n_clusters = 3, n_init= 10, max_iter =300, 
             random_state =0)
#Return a label for each data point based on nummber of cluster 
y_kmeans = kmeans.fit_predict(x)
```
